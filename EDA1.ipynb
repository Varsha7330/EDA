HISTOGRAM:
A histogram is a bar chart that shows how often certain ranges of values appear in the data. For this dataset:

A histogram of "Rate (WB)" could show how many locations have internet penetration rates in specific ranges (e.g., 10-20%, 20-30%, etc.).

Similarly, for "Users (CIA)," it can display the frequency of locations grouped by the number of internet users.


This helps to identify patterns like whether most locations have high or low internet penetration or user counts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'path_to_your_file.csv'  # Replace with your dataset's path
data = pd.read_csv("/content/internet_users.csv")

# Drop rows with missing values in the 'Users (CIA)' column
data = data.dropna(subset=['Users (CIA)'])

# Plot the histogram for 'Users (CIA)'
plt.figure(figsize=(12, 6))
sns.histplot(data['Users (CIA)'], bins=30, kde=True, color='purple')
plt.title('Histogram of Internet Users by Location')
plt.xlabel('Number of Users')
plt.ylabel('Frequency')
plt.show()

# Bar plot for Top 10 Locations by Internet Users
top_10_locations = data.nlargest(10, 'Users (CIA)')

plt.figure(figsize=(12, 6))
sns.barplot(x='Users (CIA)', y='Location', data=top_10_locations, palette='viridis')
plt.title('Top 10 Locations by Number of Internet Users')
plt.xlabel('Number of Users')
plt.ylabel('Location')
plt.show()
```

HEATMAP:
A heatmap is a visual representation of data where values are depicted by color. It’s commonly used to show relationships, patterns, or correlations in a dataset.
1. Columns Used:
"Rate (ITU)" (Internet penetration rate)

"Year.1" (Year of rate measurement)

"Users (CIA)" (Number of internet users)
2. Correlation Matrix: This calculates the correlation between the numerical columns.
3. Heatmap: Visualizes the correlations with a color scale.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'path_to_your_file.csv'  # Replace with your dataset's path
data = pd.read_csv("/content/internet_users.csv")

# Select relevant columns and drop rows with missing values
heatmap_data = data[['Rate (ITU)', 'Year.1', 'Users (CIA)']].dropna()

# Compute the correlation matrix
correlation_matrix = heatmap_data.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

PAIR PLOT:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset

data = pd.read_csv("/content/internet_users.csv")

# Select relevant numerical columns for the pair plot
selected_columns = ['Rate (WB)', 'Rate (ITU)', 'Users (CIA)', 'Year.1']
pair_plot_data = data[selected_columns].dropna()  # Drop rows with missing values

# Create the pair plot
sns.pairplot(pair_plot_data, diag_kind='kde', corner=True, palette='coolwarm')

# Show the plot
plt.suptitle('Pair Plot of Internet Usage Data', y=1.02)
plt.show()
```

Outliers are extreme values that deviate significantly from the rest of the data. Removing outliers helps in improving the accuracy of analyses and predictions.

```python
import pandas as pd

# Load the dataset

data = pd.read_csv("/content/internet_users.csv")

# Define a function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)  # First quartile
        Q3 = df[column].quantile(0.75)  # Third quartile
        IQR = Q3 - Q1                   # Interquartile range
        lower_bound = Q1 - 1.5 * IQR    # Lower bound
        upper_bound = Q3 + 1.5 * IQR    # Upper bound
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Columns where outliers need to be removed
columns_to_check = ['Rate (WB)', 'Rate (ITU)', 'Users (CIA)']

# Remove outliers
cleaned_data = remove_outliers_iqr(data, columns_to_check)

# Check the shape of the dataset after removing outliers
print("Original data shape:", data.shape)
print("Cleaned data shape:", cleaned_data.shape)
```

Insights:
After removing outliers, the data becomes cleaner and more reliable for analysis.
